```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
import seaborn as sns
from sklearn import metrics
from sklearn.svm import LinearSVC
from sklearn.model_selection import cross_val_score
import statistics
```

```python
#Importar la data que vamos a usar
data = pd.read_csv("C:\\Users\\Juan\\Desktop\\Intro_CD\\INTRO_DC_VS\\dataset1.csv")
data.reset_index(inplace=True)
data.columns = ['X1', 'X2', 'y']
data.head(10)
```

```python
# Voy a crear 1 copia del dataframe original
df = data.copy()
df.head() 
```

```python
X1=df.iloc[:,0]
X2=df.iloc[:,1]
X=np.column_stack((X1,X2))
y=df.iloc[:,2]

plt.scatter(X1[y==1], X2[y==1], c='g', marker = '+', label='1')
plt.scatter(X1[y==-1], X2[y==-1], c='b', marker = 'o', label='-1', s=10)
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.legend(bbox_to_anchor=(1.15,1.15), loc='upper right', fancybox=True, framealpha=1, fontsize=10)
# plt.show()
plt.savefig('Figure_1.png')

plt.show()
```

# Todo sobre lo relacionado al uso del modelo de Regresión Logística

```python
LR=LogisticRegression()
LR.fit(x_train, y_train)
print('The slopes are: ',LR.coef_[0])
print('The intercept is: ',LR.intercept_)
predictions_LR = LR.predict(x_train)
score = LR.score(x_train, y_train)
print('The score is: ',score)
```

Crearemos la función que utilizaremos para medir la curva de aprendizaje de un modelo, iniciando con el modelo de Regresión Logística

```python
from sklearn.model_selection import learning_curve

# Función para visualizar la curva de aprendizaje
def plot_learning_curve(model, X, y):
    train_sizes, train_scores, test_scores = learning_curve(model, X, y, cv=5, n_jobs=-1, train_sizes=np.linspace(0.1, 1.0, 10))

    # Media y desviación estándar
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)

    plt.figure(figsize=(8, 6))
    plt.plot(train_sizes, train_mean, 'o-', color="r", label="Training score")
    plt.plot(train_sizes, test_mean, 'o-', color="g", label="Cross-validation score")

    # Sombra de desviación estándar
    plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, color="r", alpha=0.1)
    plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, color="g", alpha=0.1)

    plt.title("Curva de aprendizaje para LinearSVC")
    plt.xlabel("Ejemplos de entrenamiento")
    plt.ylabel("Score")
    plt.legend(loc="best")
    plt.grid()
    plt.show()

# Visualizar curva de aprendizaje para el modelo de Regresión Logística
plot_learning_curve(LR, x_train, y_train)
```

Como se nos esta solicitando probar con distintos valores de C, vamos a crear un bucle para obtener todas las respuestas en una sola corrida. Agregaremos al bucle el cálculo de las varianzas, sesgo, perdida esperada.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import LinearSVC
from sklearn import metrics

# Define la lista de valores de C
C_values = [0.001, 0.01, 0.1, 1, 10, 100]

# Itera sobre cada valor de C
for C in C_values:
    Linear_Cl = LinearSVC(C=C)
    Model_Linear_Cl = Linear_Cl.fit(x_train,y_train)
    print(f"For C = {C}:")
    print("Los coeficientes son:", Model_Linear_Cl.coef_[0])
    print("El intercepto es:", Model_Linear_Cl.intercept_)
    predictions_L_SVC = Model_Linear_Cl.predict(x_train)
    score_SVC = Model_Linear_Cl.score(x_train, y_train)
    print("El puntaje es:", score_SVC)
    cm = metrics.confusion_matrix(y_train, predictions_L_SVC)
    print(cm)
    cr = metrics.classification_report(y_train, predictions_L_SVC)
    print(cr)
    print("\n")

    # Calcular sesgo y varianza
    y_train_mean = np.mean(y_train)
    bias = np.mean(predictions_L_SVC) - y_train_mean
    variance = np.var(predictions_L_SVC)

    # Pérdida esperada
    expected_loss = bias**2 + variance
    print(f"Sesgo: {bias:.4f}")
    print(f"Varianza: {variance:.4f}")
    print(f"Pérdida esperada: {expected_loss:.4f}")
    print("\n")
    
    # Curva de Aprendizaje
    plot_learning_curve(Model_Linear_Cl, x_train, y_train)
    print("\n" + "="*50 + "\n")
```
